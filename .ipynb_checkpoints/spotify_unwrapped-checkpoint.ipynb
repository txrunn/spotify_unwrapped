```python
# Importing core libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting display options for Pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)

# Load updated dataset from Module 1 including the newer months
df = pd.read_csv('recently_listened_spotify_5-29_12-14.csv')
df.columns = ['Timestamp', 'Track Name', 'Artist Name', 'Track ID', 'Track URL']

# Display the first few rows
df.head()
```

## Initial Data Exploration
This section summarizes the basic characteristics of the dataset to identify potential cleaning and transformation needs.

```python
# Check dataset info
df.info()

# Summarize numerical and categorical columns
df.describe(include='all')

# Check for missing values
missing_values = df.isnull().sum()
print(f"Missing Values:\n{missing_values}")
```

## Data Cleaning
### Key Cleaning Steps:
1. Remove duplicates.
2. Parse `Timestamp` into datetime format for time-based analysis.
3. Handle missing or invalid data entries.

```python
# Remove duplicates
df = df.drop_duplicates()

# Parse timestamps with an explicit format
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# Drop rows with invalid timestamps
df = df[df['Timestamp'].notna()]

# Reset index after cleaning
df.reset_index(drop=True, inplace=True)

# Display summary of cleaned data
print(f"Cleaned Data Info:\n{df.info()}")
```

## Exploratory Data Analysis (EDA)
### Objectives:
- Summarize the dataset to understand its structure and patterns.
- Identify trends in listening behavior, including popular artists, tracks, and peak listening times.
- Visualize the distribution of key features like timestamps and track popularity.

```python
# Imports for visualizations
import matplotlib.pyplot as plt
import seaborn as sns

# Set a consistent theme for visualizations
sns.set_theme(style="whitegrid")

# Top 10 most played artists
top_artists = df['Artist Name'].value_counts().head(10)
print("Top 10 Artists:\n", top_artists)

# Plotting top artists
plt.figure(figsize=(10, 6))
sns.barplot(x=top_artists.values, y=top_artists.index, palette="viridis")
plt.title("Top 10 Most Played Artists")
plt.xlabel("Number of Plays")
plt.ylabel("Artist Name")
plt.show()
```

## Time-Based Analysis
### Objectives:
- Analyze listening trends over time (hourly, daily, monthly).
- Prepare the dataset for time-series analysis using Kats.
- Visualize time-of-day listening patterns.

```python
# Extracting time-related features
df['Hour'] = df['Timestamp'].dt.hour
df['Day of Week'] = df['Timestamp'].dt.day_name()
df['Month'] = df['Timestamp'].dt.month_name()

# Hourly listening pattern
hourly_counts = df['Hour'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x=hourly_counts.index, y=hourly_counts.values, marker='o', color="blue")
plt.title("Hourly Listening Pattern")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Tracks Played")
plt.xticks(range(0, 24))
plt.grid()
plt.show()

# Weekly listening pattern
weekly_counts = df['Day of Week'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=weekly_counts.index, y=weekly_counts.values, palette="magma")
plt.title("Weekly Listening Pattern")
plt.xlabel("Day of Week")
plt.ylabel("Number of Tracks Played")
plt.show()
```

## Advanced Time-Series Analysis with Kats
### Objectives:
- Apply time-series analysis to detect seasonality and trends in listening behavior.
- Forecast future listening habits using models like Prophet.

```python
from kats.consts import TimeSeriesData
from kats.models.prophet import ProphetModel, ProphetParams

# Prepare time-series data: Aggregating daily listening counts
daily_counts = df.set_index('Timestamp').resample('D').size().reset_index(name='Track Plays')

# Convert to Kats TimeSeriesData object
ts_data = TimeSeriesData(daily_counts)

# Fit a Prophet model
params = ProphetParams(seasonality_mode='multiplicative')
model = ProphetModel(data=ts_data, params=params)
model.fit()

# Forecast future trends (e.g., next 30 days)
forecast = model.predict(steps=30, freq='D')

# Visualize the forecast
model.plot()
plt.title("Forecast of Daily Track Plays (Next 30 Days)")
plt.show()
```

```python
## Data Enrichment with Spotify API
### Objectives:
- Enhance the dataset with track-level acoustic features like tempo, danceability, and energy.
- Utilize these features to explore genre/mood patterns and cluster listening habits.
```

```python
# Spotify API imports
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Spotify API credentials
SPOTIFY_CLIENT_ID = 'your_client_id'
SPOTIFY_CLIENT_SECRET = 'your_client_secret'

# Authenticate with Spotify API
auth_manager = SpotifyClientCredentials(client_id=SPOTIFY_CLIENT_ID, client_secret=SPOTIFY_CLIENT_SECRET)
sp = spotipy.Spotify(auth_manager=auth_manager)

# Function to fetch track audio features
def fetch_audio_features(track_id):
    try:
        features = sp.audio_features(track_id)[0]
        return features
    except Exception as e:
        print(f"Error fetching features for track {track_id}: {e}")
        return None

# Apply function to Track IDs
df['Audio Features'] = df['Track ID'].apply(fetch_audio_features)

# Expand audio features into individual columns
audio_features_df = df['Audio Features'].apply(pd.Series)
df = pd.concat([df, audio_features_df], axis=1)

# Save enriched dataset for further analysis
df.to_csv('enriched_spotify_data.csv', index=False)
```

## Visualizing Acoustic Features
### Objectives:
- Explore relationships between acoustic features like energy, tempo, and danceability.
- Identify clusters of tracks based on acoustic properties.

```python
# Scatterplot of danceability vs. energy
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='danceability', y='energy', hue='valence', palette='coolwarm')
plt.title("Danceability vs. Energy (Colored by Valence)")
plt.xlabel("Danceability")
plt.ylabel("Energy")
plt.show()
```
